```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data=[[400.,5355.,3124.,3692.],[5922.,5320.,3036.,2653.],[6800.,6500.,4500.,3670.],[8700.,11000.,5000.,4000.]]
x = np.arange(4)
color_list = ['b', 'g', 'r','y']
plt.figure(figsize=(8,4))
positions = (0,1,2,3)
labels = ("2000","2005","2011","2016")
plt.xticks(positions, labels)
gap = 0.8 / len(data)
for i, row in enumerate(data):
    X = np.arange(len(row))
    plt.bar(X + i * gap, row,
    width = gap,
    color = color_list[i % len(color_list)])
plt.bar(X + 0.00, data[0], color = 'b', width = 0.10)
plt.bar(X + 0.25, data[1], color = 'g', width = 0.10)
plt.bar(X + 0.50, data[2], color = 'r', width = 0.10)
plt.bar(X + 0.75, data[3], color = 'y', width = 0.10)
plt.legend(["Delhi", "Mumbai","chennai","kolkata"], loc ="upper right")
plt.ylabel("TPD")
plt.xlabel("Years")
plt.title("Trends of Waste generation data of Metropolitan areas")
plt.show()
```

```python
df= pd.read_excel(r"E:\IIRS-WORKS\Module 4\new_project\trends.xlsx")
names = []
for i in df['Population in millions']: names.append(i)
values2001 = []
for i in df['2001(cpcb)']: values2001.append(i)
values2018 = []
for i in df['2018(cpcb)']: values2018.append(i)
```

```python
r = np.arange(len(values2001))

plt.figure(figsize=(20,8))
plt.bar(r,
        values2001, color='b', width=0.2)
plt.bar(r, values2018, bottom=values2001,color='r', width=0.2)

plt.xticks([r for r in range(len(values2001))],names)

plt.xlabel("Population in millions",  fontweight='bold')
plt.ylabel("Waste Generation per kg perday",fontweight='bold')
plt.title('Per capita waste generation range comparision in urban area)', loc ='center', fontweight='bold')
plt.legend(["CPCB 2001","CPCB 2018"])
plt.savefig('comaprison.jpeg')
```
